# Check the number of zero values in the dataset:-

```python
print("No of zero value in Glucose",data[data["Glucose"] == 0].shape[0])
print("No of zero value in BloodPressure",data[data["BloodPressure"] == 0].shape[0])
print("No of zero value in SkinThickness",data[data["SkinThickness"] == 0].shape[0])
print("No of zero value in Insulin",data[data["Insulin"] == 0].shape[0])
print("No of zero value in BMI",data[data["BMI"] == 0].shape[0])
```

# Replace the number of zero values with mean of that columns:-

```python
data["Glucose"] = data["Glucose"].replace(0,data["Glucose"].mean())
data["BloodPressure"] = data["BloodPressure"].replace(0,data["BloodPressure"].mean())
data["SkinThickness"] = data["SkinThickness"].replace(0,data["SkinThickness"].mean())
data["Insulin"] = data["Insulin"].replace(0,data["Insulin"].mean())
data["BMI"] = data["BMI"].replace(0,data["BMI"].mean())
```

# Apply an appropriate classification algorithm to build a model:-

```python
x = data.drop(["Outcome"],axis = 1)
y = data["Outcome"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
sc.fit(x)
```

```python
x = sc.transform(x)
```

```python
x_train.shape
```

```python
x_test.shape
```

1.Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr_pred = lr.predict(x_test)
lr_pred
```

2.KNeighborsClassifier(KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
knn_pred = knn.predict(x_test)
knn_pred
```

3.Naive-Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
gn = GaussianNB()
```

```python
gn.fit(x_train,y_train)
```

```python
gn_pred = gn.predict(x_test)
gn_pred
```

4.Support Vector Machine(SVM)

```python
from sklearn.svm import SVC
svc = SVC()
```

```python
svc.fit(x_train,y_train)
```

```python
svc_pred = svc.predict(x_test)
svc_pred
```

5.Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dc = DecisionTreeClassifier()
```

```python
dc.fit(x_train,y_train)
```

```python
dc_pred = dc.predict(x_test)
dc_pred
```

6.Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(criterion = "entropy")
```

```python
rf.fit(x_train,y_train)
```

```python
rf_pred = rf.predict(x_test)
rf_pred
```

# Model Evaluation:-

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
# Check the Accuracy Score
print("Accuracy Score of Logistic Regression :",accuracy_score(y_test,lr_pred)*100)
print("Accuracy Score of KNeighborsClassifier :",accuracy_score(y_test,knn_pred)*100)
print("Accuracy Score of Naive-Bayes Classifier :",accuracy_score(y_test,gn_pred)*100)
print("Accuracy Score of Support Vector Classifier :",accuracy_score(y_test,svc_pred)*100)
print("Accuracy Score of Decision Tree Classifier :",accuracy_score(y_test,dc_pred)*100)
print("Accuracy Score of Random Forest Classifier :",accuracy_score(y_test,rf_pred)*100)
```

```python
# Making the confusion matrix of various model
print("Confusion Matrix of Logistic Regression :\n\n",confusion_matrix(y_test,lr_pred)*100)
print("Confusion Matrix of KNeighborsClassifier :\n",confusion_matrix(y_test,knn_pred)*100)
print("Confusion Matrix of Naive-Bayes Classifier :\n",confusion_matrix(y_test,gn_pred)*100)
print("Confusion Matrix of Support Vector Classifier :\n",confusion_matrix(y_test,svc_pred)*100)
print("Confusion Matrix of Decision Tree Classifier :\n",confusion_matrix(y_test,dc_pred)*100)
print("Confusion Matrix of Random Forest Classifier :\n",confusion_matrix(y_test,rf_pred)*100)
```

# Create a classification report by analyzing sensitivity, specificity, AUC (ROC curve):-

```python
print("Classificaton Report of Logistic Regression :\n",classification_report(y_test,lr_pred))
```
