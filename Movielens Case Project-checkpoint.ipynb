# Import the libraries:-

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

# Import three datasets:-

```python
ratings =  pd.read_csv("ratings.dat",sep = "::",encoding = "unicode_escape",header = None)
ratings.head()
```

```python
ratings.columns = ["UserID","MovieID","Rating","Timestamp"]
ratings.head() 
```

```python
users = pd.read_csv("users.dat",sep = "::",encoding = "unicode_escape",header = None)
users.head()
```

```python
users.columns = ["UserID","Gender","Age","Occupation","Zip-code"]
users.head()
```

```python
movies = pd.read_csv("movies.dat",sep = "::",encoding = "unicode_escape",header = None)
movies.head()
```

```python
movies.columns = ["MovieID","Title","Genres"]
movies.head()
```

```python
combine = movies.merge(ratings,on = "MovieID",how = "inner")
combine.head()
```

```python
combine1 = combine.merge(users,on = "UserID",how = "inner")
combine1.head()
```

# Feature Engineering:-

# Find out all the unique genres (Hint: split the data in column genre making a list and then process the data to find out only the unique categories of genres)

```python
Genres = combine1["Genres"].str.split("|")
Genres
```

```python
list_Genres = set()
for genre in Genres:
    list_Genres = list_Genres.union(set(genre))
```

```python
list_Genres
```

# Create a separate column for each genre category with a one-hot encoding ( 1 and 0) whether or not the movie belongs to that genre.:-

```python
RatingOneHot = combine1["Genres"].str.get_dummies("|")
RatingOneHot.head()
```

```python
dfMaster = pd.concat([combine1,RatingOneHot],axis = 1)
dfMaster.head()
```

```python
dfMaster[["title","Year"]] = dfMaster.Title.str.extract("(.)\s\((.\d+)",expand = True)
```

```python
dfMaster = dfMaster.drop(columns = ["title"])
dfMaster.head()
```

```python
dfMaster["Year"] = dfMaster.Year.astype(int)
```

```python
dfMaster["Movie_Age"] = 2000 - dfMaster.Year
dfMaster.head()
```

```python
dfMaster["Gender"] = dfMaster["Gender"].replace({"F":0,"M":1})
dfMaster.head()
```

```python
sns.countplot(x = "Gender",data = dfMaster)
```

# Develop an appropriate model to predict the movie ratings:-

```python
x = dfMaster.drop(["Rating","Genres"],axis = 1)
y = dfMaster["Rating"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier() 
```

```python
model.fit(x_train,y_train)
```

```python
print(model.score(x_train,y_train)) 
```

```python
print(model.score(x_test,y_test))
```
