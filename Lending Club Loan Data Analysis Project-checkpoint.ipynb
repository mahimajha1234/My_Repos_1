# Feature Transformation:-

```python
# Handling imbalanced data
not_fully_paid_0 = df[df["not.fully.paid"] == 0]
not_fully_paid_1 = df[df["not.fully.paid"] == 1]
```

```python
not_fully_paid_0.shape
```

```python
not_fully_paid_1.shape
```

```python
#resample
from sklearn.utils import resample
df_minor_upsample = resample(not_fully_paid_1,replace=True,n_samples=8045)
```

```python
new_df = pd.concat([not_fully_paid_0,df_minor_upsample])
```

```python
# suffeling the data
from sklearn.utils import shuffle
new_df = shuffle(new_df)
```

# Create a deep learning model using Keras with Tensorflow backend:-

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# create the architecture
# 2 ANN layer
model = Sequential()
model.add(Dense(19,activation = "relu",input_shape = [9]))
model.add(Dropout(0.20))

model.add(Dense(10,activation = "relu"))
model.add(Dropout(0.20))

# output layer
model.add(Dense(1,activation = "sigmoid"))
```

```python
model.summary()
```

```python
# compile the model
model.compile(loss = "binary_crossentropy",optimizer = "adam",metrics = ["accuracy"])
```

```python
early_stop = EarlyStopping(monitor = "val_loss",min_delta = 0.01,mode = "min",patience = 10,verbose = 1)
```

```python
# fit the model
history = model.fit(x_train,y_train,epochs = 50,batch_size = 256,validation_data = (x_test,y_test),callbacks = [early_stop])
```

```python
history = model.fit(x_train,y_train,epochs = 50,batch_size = 256,validation_data = (x_test,y_test))
```

```python
model.evaluate(x_test,y_test)
```

```python
# hyperparameter tuning in keras
import keras_tuner as kt
import tensorflow
```

```python
def build_model(hp):
    model = Sequential()
    
    # first hidden layer
    model.add(Dense(units = hp.Int("units",min_value = 32,max_value = 1024,step = 16),
                   activation = hp.Choice("activation",["relu","tanh"]),input_shape = [9]))
    
    model.add(BatchNormalization())
    model.add(Dropout(hp.Float("rate",min_value = 0.1,max_value = 0.5,step = 0.1)))
            
    # second hidden layer
    model.add(Dense(units = hp.Int("units",min_value = 32,max_value = 1024,step = 16),
                   activation = hp.Choice("activation",["relu","tanh"])))
    
    model.add(BatchNormalization())
    model.add(Dropout(hp.Float("rate",min_value = 0.1,max_value = 0.5,step = 0.1)))
              
    # third hidden layer
    model.add(Dense(units = hp.Int("units",min_value = 32,max_value = 1024,step = 16),
                   activation = hp.Choice("activation",["relu","tanh"])))
    
    model.add(BatchNormalization())
    model.add(Dropout(hp.Float("rate",min_value = 0.1,max_value = 0.5,step = 0.1)))
             
    #output layer
    model.add(Dense(1,activation = "sigmoid"))
              
    learning_rate = hp.Float("learning_rate",min_value = 0.001,max_value = 0.5,step = 0.01)
    
    model.compile(loss = "binary_crossentropy",optimizer = tensorflow.keras.optimizers.Adam(learning_rate),
                 metrics = ["accuracy"])
    return model
              
```

```python
build_model(kt.HyperParameters())
```

```python
rtuner = kt.RandomSearch(hypermodel = build_model,objective = "val_accuracy",max_trials = 3)
```

```python
rtuner.search(x_train,y_train,epochs = 50,validation_data = (x_test,y_test),verbose = 2)
```
